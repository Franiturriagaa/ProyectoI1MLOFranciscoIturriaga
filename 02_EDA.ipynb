## Reconocimiento de Datos

```python
# Leo el archivo .parquet
df = pd.read_parquet('dataset_transformado.parquet')
# Reviso la estructura del DataFrame
print(df.info())
```

```python
# Examino estadisticas descriptivas de columnas numericas
print(df.describe())
```

## Dinamica de Datos

```python
# Distribucion de variables numericas
plt.figure(figsize=(3, 3))
sns.histplot(df['vote_average'], kde=True)
plt.title('Distribución del promedio votos')
plt.show()
```

## Relacion entre variables

```python
plt.figure(figsize=(20, 15))
sns.scatterplot(data=df, x='release_year', y='budget')
plt.title('Relación entre el año de lanzamiento y el costo de la pelicula')
plt.show()
```
